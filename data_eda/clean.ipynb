## 2. Lọc các cột

Giữ lại các cột cần thiết

```python
job_field_mapping = {
    "job_id": "mã công việc",
    "job_title": "tên công việc",
    "emp_id": "mã công ty",
    "emp_name": "tên công ty",
    "job_salary_unit": "đơn vị tiền tệ",
    "location_name": "địa điểm",
    "job_is_urgent_job": "khẩn cấp",
    "job_competition": "lương cạnh tranh",
    "job_eoc": "công ty yêu thích",
    "job_last_date": "hạn nộp",
    "date_view": "ngày cập nhật",
    "job_experience": "kinh nghiệm từ (năm)",
    "job_to_experience": "kinh nghiệm đến (năm)",
    "benefit_name_vn": "phúc lợi",
    "industries": "ngành",
    "job_from_salary": "lương từ",
    "job_to_salary": "lương đến",
    "job_premium_icon_item": "premium",
}

# Loc cac cot ton tai trong du lieu goc
filter_cols = [c for c in job_field_mapping.keys() if c in df_raw.columns]

# if cot_thieu:
#     print(f'Cac cot khong tim thay trong du lieu goc: {cot_thieu}')

# Tao DataFrame moi chi voi cac cot can thiet
df = df_raw[filter_cols].copy()

print(df.shape)
display(df.head().T)
```

## 3. Đổi tên cột sang tiếng Việt

```python

new_cols = {k: v for k, v in job_field_mapping.items() if k in filter_cols}
df = df.rename(columns=new_cols)

for cot in df.columns:
    print(f'  - {cot}')

df.to_csv('../dataset/careerviet_all_jobs_renamed.csv', index=False, encoding='utf-8-sig')

df.head()
```

## 4. Kiểm tra dữ liệu

```python
import pandas as pd
import numpy as np

# Tạo bản sao để thao tác an toàn
df_processed = df.copy()

# 1. Chuyển đổi các chuỗi mang ý nghĩa "trống" thành giá trị NaN thực sự của Pandas
# Regex này sẽ quét và thay thế:
# - Chuỗi "[]" hoặc có chứa khoảng trắng bên trong như "[ ]"
# - Chuỗi "null" (nếu có do lỗi JSON parsing)
# - Chuỗi rỗng "" hoặc chỉ chứa dấu cách
df_processed = df_processed.replace(r'^\s*\[\s*\]\s*$|^\s*null\s*$|^\s*$', np.nan, regex=True)

# 2. Tính số lượng thiếu (lúc này isnull() đã nhận diện được chính xác)
missing_values_count = df_processed.isnull().sum()

# 3. Tính tỉ lệ phần trăm (%) thiếu thông tin
missing_ratio_percent = df_processed.isnull().mean() * 100

# 4. Gộp kết quả thành bảng cho dễ nhìn
missing_info = pd.DataFrame({
    'Số lượng thiếu': missing_values_count,
    'Tỉ lệ thiếu (%)': missing_ratio_percent
})

# Lọc ra các cột có dữ liệu thiếu (Tỉ lệ > 0) và sắp xếp giảm dần
missing_info = missing_info[missing_info['Số lượng thiếu'] > 0].sort_values(by='Tỉ lệ thiếu (%)', ascending=False)

print(missing_info)
```

## 5. Explode các trường có nhiều giá trị

```python
def parse_list_field(value):
    """
    Chuyen doi gia tri dang chuoi danh sach thanh list Python.
    Vi du: '["Ho Chi Minh", " Binh Duong"]' -> ['Ho Chi Minh', 'Binh Duong']
    """
    if pd.isna(value):
        return []
    
    value_str = str(value).strip()
    
    # Truong hop chuoi rong hoac khong phai danh sach
    if value_str == '' or value_str == '[]':
        return []
    
    try:
        # Thu parse bang ast.literal_eval
        parsed = ast.literal_eval(value_str)
        if isinstance(parsed, list):
            # Loai bo khoang trang thua o dau va cuoi moi phan tu
            return [str(item).strip() for item in parsed if str(item).strip()]
        else:
            return [str(parsed).strip()]
    except (ValueError, SyntaxError):
        # Neu khong parse duoc, giu nguyen gia tri goc
        return [value_str.strip()]
```

```python
# Danh sach cac cot can tach (co dang danh sach nhieu gia tri)
multi_value_columns = ['địa điểm', 'phúc lợi', 'ngành']
```

```python
# Parse cac truong dang danh sach
df_clean = df.copy()

for cot in multi_value_columns:
    df_clean[cot] = df_clean[cot].apply(parse_list_field)

# Xem ket qua parse
print(df_clean['địa điểm'].head().tolist())
```

```python
# Tạo một dictionary để chứa các DataFrame mới
# Ví dụ: dataframes_exploded['industries'] sẽ trả về DataFrame chỉ explode cột industries
dataframes_exploded = {}

print(f'Số dòng ban đầu của df_clean: {len(df_clean)}\n')

for cot in multi_value_columns:
    # 1. Tạo một bản sao từ DataFrame "gốc" (df_clean) cho mỗi vòng lặp
    df_temp = df_clean.copy()
    
    # 2. Xử lý các list rỗng (thay bằng mảng chứa chuỗi rỗng để không mất dòng)
    df_temp[cot] = df_temp[cot].apply(lambda x: x if isinstance(x, list) and len(x) > 0 else [''])
    
    # 3. Tách dòng (explode)
    df_temp = df_temp.explode(cot, ignore_index=True)
    
    # 4. Xóa khoảng trắng dư thừa
    df_temp[cot] = df_temp[cot].astype(str).str.strip()
    
    # 5. Lưu DataFrame mới vào dictionary
    dataframes_exploded[cot] = df_temp
    
    print(f'Done explode cột: {cot}')

df_mapping_industries = pd.read_csv('../dataset/careerviet_industries.csv')
mapping_dict = dict(zip(df_mapping_industries['industry_id'].astype(str), df_mapping_industries['industry_name_vn']))
# Ngành ID 50 bị thiếu bởi CareerViet ẩn đi
mapping_dict["50"] = "Lâm nghiệp"

for cot, df_exploded in dataframes_exploded.items():
    # print(f'\n=== DataFrame sau khi explode cột: {cot} ===')
    # print(df_exploded.head())
    
    if cot == "ngành":
        # Thay thế mã ngành bằng tên ngành
        df_test = df_exploded[cot].copy()
        
        # 1. Tìm và in ra các giá trị không có trong từ điển (mapping_dict)
        unmapped_mask = ~df_test.astype(str).isin(mapping_dict.keys())
        unmapped_values = df_test[unmapped_mask].unique()
        
        if len(unmapped_values) > 0:
            print(f"CẢNH BÁO: Có {len(unmapped_values)} giá trị không map được, bao gồm: {unmapped_values}")
        else:
            print("Tuyệt vời! Tất cả các giá trị 'ngành' đều map thành công.")
        df_exploded[cot] = df_exploded[cot].map(mapping_dict).fillna(df_exploded[cot])
    
    df_exploded.to_csv(f'../dataset/exploded_{cot}.csv', index=False)
```

```python
df_combined = df_clean.copy()
print(f'Số dòng ban đầu: {len(df_combined)}')

for cot in multi_value_columns:
    if cot == "phúc lợi":
        continue
    
    # 1. Xử lý list rỗng cho cột hiện tại
    df_combined[cot] = df_combined[cot].apply(lambda x: x if isinstance(x, list) and len(x) > 0 else [''])
    
    # 2. Explode tuần tự đè lên chính df_combined
    df_combined = df_combined.explode(cot, ignore_index=True)
    
    # 3. Clean khoảng trắng
    df_combined[cot] = df_combined[cot].astype(str).str.strip()
    
    # 4. Map tên ngành ngay nếu cột hiện tại là "ngành"
    if cot == "ngành":
        df_combined[cot] = df_combined[cot].map(mapping_dict).fillna(df_combined[cot])
        
    print(f'Done explode cột: {cot}. Số dòng hiện tại: {len(df_combined)}')

# Xuất ra 1 file duy nhất
df_combined.to_csv('../dataset/exploded_all_combined.csv', index=False)
print('\nĐã lưu toàn bộ dữ liệu vào file exploded_all_combined.csv')
```
